# Введение

В задачах 3, 4 были реализованы 2 алгоритма для выполнения регулярных запросов к графам. Первый алгоритм представляет из себя пересечение двух автоматов при помощи произведения Кронокера и последующее транзиктивное замыкание его матрицы смежности, то есть поиск путей сразу от всех ко всем, второй –– представляет синхронный обход в ширину двух автоматов с возможностью обходить по отдельности так, если бы фронт состоял из одной стартовой вершины и сразу из всего стартового множества, то есть фронт –– множество всех стартовые вершин.

*алгоритмом номер 2(фронт –– множество) будем называть BfsSet*

*алгоритмом номер 2(фронт –– 1 стартовая вершина) будем называть BfsMultipleSource*

## Целью данной работы является:
* Определить какое представление разреженных матриц и векторов лучше подходит для каждого алгоритма
* Определить начиная с какого размера стартового множества выгоднее решать задачу алгоритмом номер 1 и алгоритмом номер 2
* Определить на сколько решение BfsSet быстрее решения BfsMultipleSource при одинаковых начальных условиях



# Описание исследуемых решений

На вход поступает регулярное выражение, граф, и множества стартовых и конечных вершин графа (состояний). Регулярное выражение и граф переводятся в соответсвующие конечные автоматы.

## Алгоритм 1 (произведение Кронокера)

У получившихся автоматов находятся булевы матрицы смежности. Далее получаем пересечение данных автоматов в виде автомата с соответсвующими булевыми матрицами смежности, полученными от произведения Кронокера двух соответсвующих матриц смежности. После, с помощью транзиктивного замыкания (умножения матрицы смежности саму на себя) находим пути достижимости от всех вершин ко всем. После фильтруем результат транзиктивного замыкания, операясь на информацию о стартовых и конечных состояниях пересечения автоматов. Итог –– множество пар достижимости от стартовой до конечной вершины.

## Алгоритм 2 (BfsSet)

Получившиеся автоматы синхронно обходим в ширину до тех пор пока можно ходить одновременно по двум автоматам по одинаковым ребрам. Синхронный обход производится путем умножения матрицы фронта на матрицу прямой суммы булевых матриц двух автоматов. Матрица фронта ––  матрица K x K+N (K –– кол-во состояний в 1-ом автомате, N –– кол-во состояний в 2-ом автомате). Первый блок KxK –– диагональная матрица, где на диагонали стоит True с индексом ixi, если индекс i –– индекс стартовой вершины в 1-ом автомате. Второй блок правее от блока KxK, его размерность –– KxN. True с индексом ixj, если i –– стартовое состояние автомата 1 и j –– стартовое состояние автомата 2.

## Алгоритм 2 (BfsMultipleSource)

Получившиеся автоматы синхронно обходим в ширину до тех пор пока можно ходить одновременно по двум автоматам по одинаковым ребрам. Синхронный обход производится путем умножения матрицы фронта на матрицу прямой суммы булевых матриц двух автоматов. Матрица фронта ––  матрица K*K x K+N (K –– кол-во состояний в 1-ом автомате, N –– кол-во состояний в 2-ом автомате). Левые блоки KxK –– матрицы, где на диагонали стоят True с индексом ixi, если индекс i –– индекс стартовой вершины в 1-ом автомате (такое True всего одно для каждой матрицы KxK). Вторые блоки правее от блока KxK, его размерность –– KxN. True с индексом ixj, если i –– стартовое состояние автомата 1 и j –– стартовое состояние автомата 2.

```python
import sys
import os
path = os.getcwd()[:-11]
if path not in sys.path: sys.path.append(os.getcwd()[:-11])
    
import random
import statistics

from timeit import timeit
from networkx import MultiDiGraph
from pyformlang.regular_expression import Regex
from cfpq_data import generate_multiple_source_percent

from project.graph_utilities import get_graph
from project.rpq import rpq
from project.rpq import MethodRpq
from project.rpq import MethodBfs
from project.rpq import MethodTensor

graph_names = ["skos", "funding", "wc", "bzip", "gzip", "pathways"] 
graphs = {graph_name: get_graph(graph_name) for graph_name in graph_names}
execution_times_number = 20

def generate_regexes(graph: MultiDiGraph) -> tuple[str, str, str]:
    graph_lables = list({el["label"] for (_, _, el) in graph.edges(data=True)})

    regex1 = f"({'+'.join(graph_lables)})*"

    regex2 = f"{random.choice(graph_lables)}"

    def regex3():
        a, b = random.choices(graph_lables, k=2)
        while b == a:
            b = random.choice(graph_lables)
        return f'{a}.{a}*.({b}+{a})'

    return regex1, regex2, regex3()

def repeat(stmt) -> list[float]:
    times = []
    for i in range(execution_times_number):
        times.append(timeit(stmt=stmt, number=1))
    return [statistics.mean(times), statistics.stdev(times)]

experiment_data = []
experiment_data_matrix_test = []
experiment_regexes = {
    graph_name: generate_regexes(graphs.get(graph_name))
    for graph_name in graphs.keys()
}

def fill_experiment_data_matrix_test(percent: float, seed: int):
    for graph_name in graph_names:
        if graph_name in ["gzip", "pathways"]:
            if percent > 1.0: continue
            else: percent = 0.2
        print(f"{graph_name} started")
        graph = graphs.get(graph_name)
        tuple_experiment_regexes = experiment_regexes.get(graph_name)
        regex = tuple_experiment_regexes[0]
        algo1csr = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed)))
        algo1csc = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed), csr=False))
        experiment_data_matrix_test.append([graph_name, "tensor", "(Union_all)*", f"{percent}%", "csr"] + algo1csr)
        experiment_data_matrix_test.append([graph_name, "tensor", "(Union_all)*", f"{percent}%", "csc"] + algo1csc)
        print(f"{graph_name} completed part 1")

    for graph_name in graph_names:
        if graph_name in ["gzip", "pathways"]:
            if percent > 1.0: continue
            else: percent = 0.2
        print(f"{graph_name} started")
        graph = graphs.get(graph_name)
        tuple_experiment_regexes = experiment_regexes.get(graph_name)
        regex = tuple_experiment_regexes[0]
        algo2csr = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed), method_rpq=MethodRpq(MethodBfs(is_multiple_source=False))))
        algo2csc = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed), method_rpq=MethodRpq(MethodBfs(is_multiple_source=False)), csr=False))
        experiment_data_matrix_test.append([graph_name, "bfs_set", "(Union_all)*", f"{percent}%", "csr"] + algo2csr)
        experiment_data_matrix_test.append([graph_name, "bfs_set", "(Union_all)*", f"{percent}%", "csc"] + algo2csc)
        print(f"{graph_name} completed part 2")

    for graph_name in graph_names:
        if graph_name in ["gzip", "pathways"]:
            if percent > 1.0: continue
            else: percent = 0.2
        print(f"{graph_name} started")
        graph = graphs.get(graph_name)
        tuple_experiment_regexes = experiment_regexes.get(graph_name)
        regex = tuple_experiment_regexes[0]
        algo3csr = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed), method_rpq=MethodRpq(MethodBfs(is_multiple_source=True))))
        algo3csc = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed), method_rpq=MethodRpq(MethodBfs(is_multiple_source=True)), csr=False))
        experiment_data_matrix_test.append([graph_name, "bfs_ms", "(Union_all)*", f"{percent}%", "csr"] + algo3csr)
        experiment_data_matrix_test.append([graph_name, "bfs_ms", "(Union_all)*", f"{percent}%", "csc"] + algo3csc)
        print(f"{graph_name} completed fully")

def fill_experiment_data(percent: float, seed: int):
    for graph_name in graph_names:
        if graph_name in ["gzip", "pathways"]:
            if percent > 1.0: continue
            else: percent = 0.2
        print(f"{graph_name} started")
        graph = graphs.get(graph_name)
        tuple_experiment_regexes = experiment_regexes.get(graph_name)
        regex = tuple_experiment_regexes[0]
        algo1 = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed)))
        algo2 = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed), method_rpq=MethodRpq(MethodBfs(is_multiple_source=False))))
        algo3 = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed), method_rpq=MethodRpq(MethodBfs(is_multiple_source=True))))
        experiment_data.append([graph_name, "tensor", "(Union_all)*", f"{percent}%"] + algo1)
        experiment_data.append([graph_name, "bfs_set", "(Union_all)*", f"{percent}%"] + algo2)
        experiment_data.append([graph_name, "bfs_ms", "(Union_all)*", f"{percent}%"] + algo3)

        regex = tuple_experiment_regexes[1]
        algo1 = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed)))
        algo2 = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed), method_rpq=MethodRpq(MethodBfs(is_multiple_source=False))))
        algo3 = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed), method_rpq=MethodRpq(MethodBfs(is_multiple_source=True))))
        experiment_data.append([graph_name, "tensor", f"{regex}", f"{percent}%"] + algo1)
        experiment_data.append([graph_name, "bfs_set", f"{regex}", f"{percent}%"] + algo2)
        experiment_data.append([graph_name, "bfs_ms", f"{regex}", f"{percent}%"] + algo3)

        regex = tuple_experiment_regexes[2]
        algo1 = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed)))
        algo2 = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed), method_rpq=MethodRpq(MethodBfs(is_multiple_source=False))))
        algo3 = repeat(lambda: rpq(regex, graph, start_states = generate_multiple_source_percent(graph, percent, seed=seed), method_rpq=MethodRpq(MethodBfs(is_multiple_source=True))))
        experiment_data.append([graph_name, "tensor", f"{regex}", f"{percent}%"] + algo1)
        experiment_data.append([graph_name, "bfs_set", f"{regex}", f"{percent}%"] + algo2)
        experiment_data.append([graph_name, "bfs_ms", f"{regex}", f"{percent}%"] + algo3)
        print(f"{graph_name} completed fully")
```

Будем использовать prettytable для того, чтобы создавать готовые таблицы:

```python
import prettytable
from prettytable import PrettyTable

def print_table(header: list[str], rows: list[list[any]]):
    table = PrettyTable()
    table.set_style(prettytable.MARKDOWN)
    table.field_names = header
    for row in rows:
        table.add_row(row)
    table.max_width = 33
    print(table)
    
def to_fixed(numObj, digits=0):
    return f"{numObj:.{digits}f}"
```
